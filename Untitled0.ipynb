```python
image_face = io.imread("/content/drive/MyDrive/Colab Notebooks/GoogleColab/istockphoto-1184216653-612x612.jpg") 
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(image_face, 1.1 , 5 )
for (x,y,w,h) in faces:   
     cv.rectangle(image_face,(x,y),(x+w,y+h),(255,0,0),2)
```
